```python
# Crear un diccionario de frecuencia de palabras
frecuencia_palabras = {}

# Crear una lista de palabras utilizando la variable letra_cancion
letra_cancion = """
Te vas a llevar todo
Lo que no hay en Hong Kong no exi—
Lo que no hay en Hong Kong es que no existe

Salimos a la calle por tabaco
Volvimos sin tabaco ni mechero
Tomamos hasta el agua del florero
Aquello fue peor que Puerto Hurraco

Me cago en las personas que se rajan
Muy antes de que se acabe la noche
Peinábamos perico con navaja
En el salpicadero de tu coche

Tengo una flor en el c*lo y un camello en Hong Kong
Tengo un cohete en el pantalón
Tengo una flor en el c*lo y una geisha en Japón
Tus banderillas en el corazón

Hagamos que parezca un accidente
Decirme que no tengo que ir a casa
Espero que esto dure para siempre
Sabemos qué es torear en peores plazas

Historias que jamás pueden contarse
Batallas que no pueden repetirse
Victorias para el que supo quedarse
¿Pa' qué cojones sirve arrepentirse?

Tengo una flor en el c*lo y un camello en Hong Kong
Tengo un cohete en el pantalón
Tengo una flor en el c*lo y una geisha en Japón
Tus banderillas en el corazón

Tengo una flor en el c*lo y un camello en Hong Kong
Tengo un cohete en el pantalón
Tengo una flor en el c*lo y una geisha en Japón
Tus banderillas en el corazón
"""

# Crear la lista de palabras
palabras = letra_cancion.split()

# Llenar el diccionario de frecuencia de palabras
for palabra in palabras:
    palabra_limpia = palabra.lower().strip(".,!?—")
    if palabra_limpia in frecuencia_palabras:
        frecuencia_palabras[palabra_limpia] += 1
    else:
        frecuencia_palabras[palabra_limpia] = 1

# Mostrar el diccionario de frecuencia de palabras
frecuencia_palabras
```

```python
# Ordenar alfabéticamente las claves del diccionario
claves_ordenadas = sorted(frecuencia_palabras.keys())

# Mostrar las claves ordenadas
claves_ordenadas
```

```python
# Crear un nuevo diccionario con las claves ordenadas alfabéticamente
frecuencia_palabras2 = {clave: frecuencia_palabras[clave] for clave in claves_ordenadas}

# Mostrar el nuevo diccionario
frecuencia_palabras2
```

```python
# Ordenar el diccionario frecuencia_palabras2 por sus valores de forma ascendente
frecuencia_palabras2_ordenado = dict(sorted(frecuencia_palabras2.items(), key=lambda item: item[1]))

# Mostrar el diccionario ordenado
frecuencia_palabras2_ordenado
```
